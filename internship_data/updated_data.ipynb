# Linear Regression Model

```python
#x parameters(independent variables)
cat=df[['OPEN','HIGH','LOW','LAST','PREVCLOSE','TOTTRDQTY','TOTTRDVAL','TOTALTRADES']]
cat.head()
```

```python
#defining x and y variables
y = df.CLOSE
x = cat
```

```python
# Split The dataset into training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
y_pred_linear=model.predict(x)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y,y_pred_linear)
print("R2 score of linear regression model:",r2)
```

# Polynomial Model

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_feature=PolynomialFeatures(degree=2)
poly_feature
```

```python
x_poly_train=poly_feature.fit_transform(x)
x_poly_test=poly_feature.fit_transform(x)
```

```python
model_poly=LinearRegression()
model_poly.fit(x_poly_train,y)
```

```python
y_pred_Poly = model_poly.predict(x_poly_test)
```

```python
from sklearn.metrics import r2_score
r2_pr = r2_score(y, y_pred_Poly)
print("R2_Score after creating Polynomial Features:", r2_pr)
```

```python
result = pd.DataFrame({ "Actual": y, 
                        "Predicted(Linear)": y_pred_linear, 
                        "Predicted": y_pred_Poly,
                      })
```

```python
result.head()
```
